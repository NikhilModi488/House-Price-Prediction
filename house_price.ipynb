```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
import pickle
import bz2
```

```python
data_frame=pd.read_csv('Train.csv')
pd.set_option('display.max_columns',None)
print(data_frame.head())
```

```python
posted_by=data_frame.POSTED_BY
posted_by=pd.get_dummies(posted_by)
# print(posted_by)
```

```python
bhk_or_rk=data_frame.BHK_OR_RK
bhk_or_rk=pd.get_dummies(bhk_or_rk)
```

```python
x=data_frame.iloc[:,:-1]
y=data_frame.iloc[:,-1:]
x=pd.concat([x,posted_by,bhk_or_rk],axis=1)
x=x.drop(['POSTED_BY','BHK_OR_RK','ADDRESS'],axis=1)
print("x: ",x,"y: ",y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=123)
```

```python
random_forest=RandomForestRegressor()
random_forest.fit(x_train,y_train)
rf_y_pred=random_forest.predict(x_test)
rf_score=random_forest.score(x_test,y_test)
print(rf_score)
```

```python
file=open('house_price.pkl','wb')
pickle.dump(random_forest,file)
file.close()
```
